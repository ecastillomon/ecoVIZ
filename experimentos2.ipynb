```python
import pandas as pd
```

```python
estaciones=pd.read_csv('data/stations.csv')
estaciones=estaciones.set_index('id')
estaciones.head()
```

```python
dia=pd.read_csv('data/12-06-19.csv')

dia=dia[:22630]
```

```python
dia=dia.join(estaciones,on='estacion',how='left')
dia=dia.sort_values(by=['hora','estacion'])
dia.head()
```

# Dejar datos listos para gif

```python
dia.groupby(by=['estacion','hora'])['n_retiro'].rolling(4,min_periods=1).sum()
```

```python
dia['n_arribo_acum']=dia.groupby(by=['hora','estacion'])['n_arribo'].rolling(4,min_periods=1).sum().reset_index(drop=True)
```
