La red de ecobicis es un sistema que brinda servicio a una enorma cantidad de usuarios, cómo tal, es un sistema complejo que cuenta con 480 estaciones y tiene cerca de 600,000 viajes al mes. El uso de la bicicleta tiene beneficios en salud y ahorra tiempos de traslado en distancias cortas a los usuarios, mientras ayuda al medio ambiente, y disminuye el tráfico en la ciudad.

## El estado de la red

La cantidad de datos generados por **ECOBICI** es inmensa ya que se genera un registro por viaje, cada uno de estos viajes cuenta con campos como:

* Edad y género del usuario del viaje.
* Estación de retiro y arribo.
* Hora de retiro y arribo.
* Entre otros.

Para empezar a entender el sistema de **ECOBICI** empezemos por visualizar como son las distribuciones de viajes a lo largo de una semana.

Es interesante ver la distinción clara entre los días laborales y los fines de semana. Entre días laborales las distribuciones parecen ser idénticas con la excepción del viernes por la tarde. En días feriados, que en la **Figura 2.** coinciden con los días de Semana Santa, se observa una distribución casi idéntica a la de los fines de semana. Los días laborales son los que tienen la mayor cantidad de viajes, estos días tienen una distribución bimodal, con los picos coincidiendo con las horas de entrada y salida de la jornada laboral. Los fines de semana el sistema llega a su máximo uso a medio día.

### Un día laboral

Los días laborales son los más activos con 2 horas pico claras en la mañana y en la tarde, viendo las gráficas anteriores podemos afirmar sin miedo que los días laborales son practicamente idénticos, es por esto que pensamos que visualizando la actividad en un día laboral puede ser suficiente para entender la mayoría de la actividad en el sistema.

El siguiente video muestra la actividad total de **ECOBICI** el día 12 de junio del 2019, se seleccionó este día por ser un día laboral cualquiera en la muestra, teniendo cuidado de no tomar días atípicos por clima u otras razones. Este día hubo 32363 viajes en total, con 3738 viajes por hora durante la máxima hora pico.

Síentete libre de pausar en las horas pico (8am y 8pm aprox.).

Tomemos el tiempo para entender el gráfico anterior y así poder extraer toda la información que nos otorga. Empecemos por visualizar a las estaciones como cada uno de los círculos. El tamaño del círculo corresponde a la suma de retiros y arribos que hubo en esa estacion por hora, siendo 140 arribos más retiros la actividad máxima por hora en una estación. Cada una de estas estaciones está coloreada en el interior y en el exterior. El color **interior** responde al balance presentado por hora, es decir, si en esa estación han habido más retiros o arribos en una hora. Un color rojo obscuro responde a sólo retiros y un azul obscuro a sólo arribos, blanco muestra un balance perfecto. El color **exterior** responde al estado de la estación, es decir, la proporción de los espacios para bicicletas ocupados por una **ECOBICI** en esa estación al corte de tiempo, un color rojo obscuro responde a una estación sin bicicletas disponibles mientras que un color blanco muestra una estación totalmente ocupada. Los viajes están definidos por las lineas rectas entre 2 estaciones en las que se muestra la cantidad de viajes por hora.

Una reproducción del video permite identificar correctamente las horas pico, además es posible ver como en las mañanas, las estaciones cercanas al transporte público como el metro polanco y metro Buenavista son los puntos de partida de la mayoría de los viajes, viajes que terminan en zonas con alta ocupación laboral como algunas partes de Polanco y la avenida Reforma. Este fenómeno se reproduce casi idéntico en la tarde sólo que en sentido opuesto, ahora las zonas con alta ocupación laboral resultan ser los puntos de partida y los puntos de llegada son intersecciones con transporte público de mayor alcance. Veamos un sólo cuadro del vídeo para la hora pico de la mañana.

La visualización destaca la saturación del sistema en su periferia. Esto sugiere que una buena parte de los usuarios del sistema están llegando de otros lugares, y que probablemente estarían interesados en una cobertura más amplia del sistema. Además si nos concentramos en el estado de las estaciones(el borde de cada círculo) es posible ver que tras cada ráfaga de llegadas y salidas, las estaciones quedan llenas y vacías pues al haber muchas llegadas estas se saturan(bordes blancos) y al haber muchas salidas estas se vacían(bordes negros). La dinámica particular de los sistemas de bicicletas compartidas sugieren hacer enfásis en una optimización del inventario mediante un rebalanceo, como se muestra en *-- Castillo ,2019.* para así evitar saturaciones y huecos.

Aproximarse a esta visualización permite generar una primera ídea bastante robusta de una posible red de ciclopistas y nuevas estaciones. Exigencias justificadas pues muchas de estas rutas ya están siendo ocupadas, apartar algunos metros en vías públicas para el flujo de estos ecociclistas y ciclistas sólo sería reconocerles y brindarles seguridad y reconocimiento por parte del gobierno. Además de favorecer sistemas de movilidad inteligentes aun con los pocos incentivos que existen para ellos. Darles escala a estas formas de movilidad permitiría mejorar muchas variables sociales que se encuentran en este momento en números rojos.

* Incremento de la salud por incremento en la actividad física.
* Incremento en la calidad del aire por disminución de transportes propulsados por hidrocarburos.
* Disminución en la congestión vial.
* Disminución en la sobrecarga de otros sistemas de transporte público, sobre todo en la famosa última milla.
* Entre otros efectos conocidos propiciados por el uso de la bicicleta.

## Conclusiones

La atención adecuada en la escala de los proyectos sociales contemporáneos es una de las variables más importantes en el éxito de estos, es por esto que creémos en la cooperación. Además estamos seguros que la búsqueda y validación de soluciones públicas debe ser un ejercicio colectivo,por esto, la totalidad de los análisis desarrollados en este proyecto pueden ser replicados, pues todos los scripts están disponibles en este [repositorio](https://github.com/est092/ecoVIZ) público. El árticulo fue diseñado en un **Jupyter Notebook** mientras que el contenido fue elaborado 50% en **Python** con gráficos de **matplotlib** y 50% en <b>R</b>  con **ggplot** por mencionar algunas de las librerías de software libre que fueron ocupadas.

La correcta interpretación de los datos permite el apalancamiento necesario para una implementación eficaz de políticas públicas, celebramos el reciente [anuncio](https://www.jefaturadegobierno.cdmx.gob.mx/comunicacion/nota/anuncia-gobierno-de-la-ciudad-de-mexico-nueva-licitacion-para-el-sistema-ecobici) emitido por la secretaría de movilidad para una nueva licitación para el sistema **ECOBICI** y esperamos que el ganador tome en cuenta la información existente.
